```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iq_data = pd.read_csv('iq.csv')
hdi_data = pd.read_csv('hdi.csv')
# economic_freedom_data = pd.read_csv('drive/MyDrive/Statistics/economic_freedom.csv')
income_data = pd.read_csv('income.csv')
```

```python
iq = []
for i in range(len(iq_data['iq'])):
  iq.append(iq_data['iq'][i])

print(iq)
```

```python
hdi = []
for i in range(len(hdi_data['hdr2019'])):
  hdi.append(hdi_data['hdr2019'][i])

print(hdi)
```

```python
income = []
for i in range(len(income_data['meanIncome'])):
  income.append(income_data['meanIncome'][i])

print(income)
```

```python
countries = []

for i in range(len(iq_data['country'])):
  countries.append(iq_data['country'][i])

for i in range(len(hdi_data['country'])):
  countries.append(hdi_data['country'][i])

for i in range(len(income_data['country'])):
  countries.append(income_data['country'][i])

countries = set(i for i in countries if countries.count(i) > 2)
print(countries)
```

```python
iq_dict = iq_data.to_dict('split')
iq_list = []

for i in range(len(iq_dict['data'])):
  if iq_dict['data'][i][1] in countries:
    iq_list.append(iq_dict['data'][i][2])
```

```python
hdi_dict = hdi_data.to_dict('split')
hdi_list = []

for i in range(len(hdi_dict['data'])):
  if hdi_dict['data'][i][0] in countries:
   hdi_list.append(hdi_dict['data'][i][1])
```

```python
income_dict = income_data.to_dict('split')
income_list = []

for i in range(len(income_dict['data'])):
  if income_dict['data'][i][0] in countries:
   income_list.append(income_dict['data'][i][2])
```

```python
# Concat all the lists

all_data = pd.DataFrame(
  {'Country': sorted(countries),
    '2019 HDI': hdi_list,
    'Average IQ': iq_list,
    'Mean Income': income_list
  })
```

```python
all_data
```



---



```python
plt.title('HDI & Income')
plt.xlabel('2019 HDI')
plt.ylabel('Mean Income')
plt.scatter(all_data['2019 HDI'], all_data['Mean Income'])
```

```python
plt.title('IQ & Income')
plt.xlabel('Average IQ')
plt.ylabel('Mean Income')
plt.scatter(all_data['Average IQ'], all_data['Mean Income'])
```

```python
x = all_data[['2019 HDI', 'Average IQ']]
y = all_data['Mean Income']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1)
```

```python
from sklearn.linear_model import PoissonRegressor

model = PoissonRegressor()
model.fit(x_train, y_train)
```

```python
# make predictions
predictions = model.predict(x_test)
```

```python
plt.hist(y_test - predictions)
```

```python
from sklearn import metrics

mse = metrics.mean_absolute_error(y_test, predictions) 
print(mse)
```
